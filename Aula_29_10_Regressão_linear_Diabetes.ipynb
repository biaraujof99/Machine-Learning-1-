```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
```

```python
from sklearn.metrics import mean_squared_error , r2_score
```

```python
X, y = datasets.load_diabetes(return_X_y=True)
#Função que signica que o valor vai retornar com dois valores
```

```python
X = X [:,np.newaxis, 2]
```

```python
plt.scatter(X, y, color="pink")
```

```python
X_treino = X[:-20]
X_teste = X[-20:]
```

```python
y_treino = y[:-20]
y_teste = y[-20:]
```

```python
linear = linear_model.LinearRegression()
```

```python
linear.fit(X_treino,y_treino)
```

```python
previsoes = linear.predict(X_teste)
```

```python
print("Coeficiente Angular:", linear.coef_)
```

```python
print("Erro quadratico médio: %.2f" % mean_squared_error(y_teste, previsoes))
```

```python
print("Coeficiente de determinação: %.2f" % r2_score(y_teste, previsoes))
```

```python
plt.scatter(X_teste, y_teste, color= "pink")
plt.plot(X_teste, previsoes, color="black", linewidth=2)
plt.show()

#Para Diabetes o modelo identificou que o IMC, apenas, não é suficente para determinar o modelo preditivo.
```
